## importamos librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, calinski_harabasz_score
from scipy.cluster.hierarchy import dendrogram, linkage

import os
os.environ["OMP_NUM_THREADS"] = "1"
```

## Cargar y explorar el conjunto de datos

```python
# Cargar el dataset
data = pd.read_csv('Mall_Customers.csv')  

# informaci√≥n del dataset
data.info()
data.describe()
```

```python
data.head(20)
```

## analisis exploratorio

```python
sns.histplot(data['Annual Income (k$)'], kde=True)
sns.histplot(data['Spending Score (1-100)'], kde=True)
```

## procesamiento de datos

```python
# Eliminar columnas irrelevantes, como ID, si no es √∫til para el an√°lisis
data = data.drop(columns=['CustomerID'])

# Convertir 'Genre' a num√©rico si se incluye en el an√°lisis
data['Gender'] = data['Gender'].map({'Male': 1, 'Female': 0})

# Escalado de datos (opcional para algunos algoritmos)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data[['Annual Income (k$)', 'Spending Score (1-100)']])
```

```python
sns.pairplot(data.dropna(), height=3,kind='scatter')
```

## seleccion de caracteristicas

### Para este ejemplo, seleccionamos Ingreso Anual y Puntaje de Gasto como caracter√≠sticas clave para entrenar los modelos.

## entrenar modelos

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="sklearn.cluster")
# Encontrar el n√∫mero √≥ptimo de clusters con el m√©todo de codo
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(scaled_data)
    sse.append(kmeans.inertia_)

plt.plot(range(1, 11), sse)
plt.xlabel('N√∫mero de Clusters')
plt.ylabel('SSE')
plt.title('M√©todo del Codo')
plt.show()

# Entrenar el modelo K-means con el n√∫mero √≥ptimo de clusters
kmeans = KMeans(n_clusters=5, random_state=0)  
kmeans.fit(scaled_data)
data['KMeans_Cluster'] = kmeans.labels_
```

## Agrupamiento Jer√°rquico

```python
# Crear un dendograma para visualizar la estructura de clusters
linked = linkage(scaled_data, method='ward')
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.show()

# Entrenar el modelo de clustering jer√°rquico
hc = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')  # Ajusta n_clusters
data['Hierarchical_Cluster'] = hc.fit_predict(scaled_data)
```

## evaluacion del modelo

```python
# Coeficiente de Silhouette para ambos modelos
silhouette_kmeans = silhouette_score(scaled_data, data['KMeans_Cluster'])
silhouette_hc = silhouette_score(scaled_data, data['Hierarchical_Cluster'])

# √çndice de Calinski-Harabasz
calinski_kmeans = calinski_harabasz_score(scaled_data, data['KMeans_Cluster'])
calinski_hc = calinski_harabasz_score(scaled_data, data['Hierarchical_Cluster'])

print(f"K-means Silhouette Score: {silhouette_kmeans}")
print(f"Hierarchical Clustering Silhouette Score: {silhouette_hc}")
print(f"K-means Calinski-Harabasz Score: {calinski_kmeans}")
print(f"Hierarchical Clustering Calinski-Harabasz Score: {calinski_hc}")
```

```python
# Visualizar los clusters de K-means
plt.figure(figsize=(10, 6))
sns.scatterplot(x=scaled_data[:, 0], y=scaled_data[:, 1], hue=data['KMeans_Cluster'], palette='viridis')
plt.title('K-means Clustering')
plt.xlabel('Ingreso Anual (Normalizado)')
plt.ylabel('Puntaje de Gasto (Normalizado)')
plt.show()

# Visualizar los clusters de Clustering Jer√°rquico
plt.figure(figsize=(10, 6))
sns.scatterplot(x=scaled_data[:, 0], y=scaled_data[:, 1], hue=data['Hierarchical_Cluster'], palette='viridis')
plt.title('Clustering Jer√°rquico')
plt.xlabel('Ingreso Anual (Normalizado)')
plt.ylabel('Puntaje de Gasto (Normalizado)')
plt.show()
```

Estos valores reflejan que los clusters tienen una cohesi√≥n interna aceptable y est√°n moderadamente separados entre s√≠. Aunque K-means tiene un desempe√±o ligeramente superior
